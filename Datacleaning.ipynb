```python
#select avg.
#fill the avg value  with mean
df['Shows'].fillna(df['Shows'].mean())
```

```python
# help us to create our outlier
df.describe()
```

```python
# calculate IQR
q1 = df["Shows"].quantile(0.25)
q3 = df["Shows"].quantile(0.75)
IQR = q3 - q1
```

```python
#define outlier boundries
lower_bound = q1 - 1.5 * IQR
upper_bound = q3 + 1.5 * IQR
```

```python
lower_bound
```

```python
upper_bound
```

```python
df[df["Shows"]<= lower_bound]
```

```python
df[df["Shows"]>= upper_bound]
```

```python
# calculatre the zscore
df['Z_score'] = np.abs(stats.zscore(df['Shows']))
```

```python
#identify outliers (typically Z-score > 3 is considered an outlier)
outliers = df[df['Z_score'] > 3]
print("Outliers:")
print(outliers)
```
